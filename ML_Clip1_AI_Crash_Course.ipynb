## หัวข้อที่ 2 — Machine Learning ทำงานอย่างไร?
**เวลาโดยประมาณ: 5 นาที**

Machine Learning เป็นแนวทางที่ให้คอมพิวเตอร์ **เรียนรู้รูปแบบ (Pattern) จากข้อมูล** แทนที่จะถูกเขียนกฎเกณฑ์ตายตัว

### เปรียบเทียบการเขียนโปรแกรมแบบปกติกับ ML

**วิธีปกติ (Rule-based Programming):**
```
กฎ + ข้อมูล → คำตอบ
```
ตัวอย่าง: เขียนโค้ดระบุว่า "ถ้า PM2.5 > 50 → อากาศไม่ดี"

**วิธี Machine Learning:**
```
ข้อมูล + คำตอบ → กฎ (Model)
```
ให้คอมพิวเตอร์หากฎเอง จากตัวอย่างที่เราให้!

### 3 ประเภทหลักของ Machine Learning

| ประเภท | วิธีการ | ตัวอย่าง |
|--------|--------|----------|
| **Supervised Learning** | มีคำตอบให้ดู (มีครูสอน) | จำแนกสแปม, ทำนายราคาบ้าน |
| **Unsupervised Learning** | ไม่มีคำตอบ (หาเอง) | จัดกลุ่มลูกค้า, หาของผิดปกติ |
| **Reinforcement Learning** | ลองผิดลองถูก (ได้รางวัล) | AI เล่นเกม, หุ่นยนต์เดิน |

## หัวข้อที่ 4 — ML Pipeline: ขั้นตอนการสร้าง AI
**เวลาโดยประมาณ: 5 นาที**

ML Pipeline คือกระบวนการทำงานมาตรฐานในการสร้างระบบ Machine Learning ตั้งแต่เริ่มต้นจนใช้งานจริง

```
1. ตั้งโจทย์  →  2. เตรียมข้อมูล  →  3. เลือก Model  →  4. Train  →  5. ทดสอบ  →  6. ใช้งาน
```

ต่อไปนี้จะลองเขียนโค้ด ML ด้วย Python โดยใช้ไลบรารี `scikit-learn` ซึ่งเป็นเครื่องมือมาตรฐานสำหรับ Machine Learning

````python
# ตัวอย่าง ML Pipeline ด้วย Python — ทำนายเพศจากส่วนสูงและน้ำหนัก
from sklearn.tree import DecisionTreeClassifier

# 1. เตรียมข้อมูล: [ส่วนสูง (cm), น้ำหนัก (kg)] → เพศ
X = [[170, 65], [180, 80], [160, 50], [155, 45],
     [175, 70], [185, 90], [150, 42], [165, 55]]
y = ["M", "M", "F", "F", "M", "M", "F", "F"]

# 2. สร้างและ Train Model
model = DecisionTreeClassifier()
model.fit(X, y)

# 3. ทำนายข้อมูลใหม่
prediction = model.predict([[172, 68]])
print(f"ส่วนสูง 172 cm น้ำหนัก 68 kg → AI ทำนาย: {prediction[0]}")

prediction2 = model.predict([[158, 48]])
print(f"ส่วนสูง 158 cm น้ำหนัก 48 kg → AI ทำนาย: {prediction2[0]}")

print("\nสรุป 3 ขั้นตอนหลัก: สร้าง → Train → ทำนาย!")
````
